```python
from policy import CustomPolicy 
from PPO_RC import PPO_RC

from stable_baselines.common.vec_env import DummyVecEnv, SubprocVecEnv
from stable_baselines.common import set_global_seeds

def get_gym_data():
    data, data_next = [], []
    env_id = "MountainCarContinuous-v0"

    env_test = gym.make(env_id)
    
    model = PPO_RC(CustomPolicy, DummyVecEnv([lambda: env_test]), n_steps=int(1024), ent_coef=0.0, nminibatches=64, noptepochs=4, \
                                    verbose=1, tensorboard_log='/home/xi/model')
    obs = env_test.reset()
    for i in range(5000):
        data.append(obs.copy())
        action, _states = model.predict(obs, deterministic=True)
        obs, rewards, dones, info = env_test.step(action)
        data_next.append(obs.copy())
#         env_test.render()
        if dones:
            env_test.reset()
    return np.asarray(data), np.asarray(data_next)
```

```python
class VariantionalAutoencoder(object):

    def __init__(self, learning_rate=1e-4, batch_size=64, n_z=16, recon_coef=1):
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.n_z = n_z
        self.recon_coef = recon_coef

        tf.reset_default_graph()
        self.build()

        self.sess = tf.InteractiveSession()
        self.sess.run(tf.global_variables_initializer())

    # Build the netowrk and the loss functions
    def build(self):
        input_dim = 2
        self.x = tf.placeholder(
            name='x', dtype=tf.float32, shape=[None, input_dim])

        # Encode
        # x -> z_mean, z_sigma -> z
        f1 = fc(self.x, 64, scope='enc_fc1', activation_fn=tf.nn.relu)
        f2 = fc(f1, 32, scope='enc_fc2', activation_fn=tf.nn.relu)
        f3 = fc(f2, 16, scope='enc_fc3', activation_fn=tf.nn.relu)
        self.z_mu = fc(f3, self.n_z, scope='enc_fc4_mu', 
                       activation_fn=None)
        self.z_log_sigma_sq = fc(f3, self.n_z, scope='enc_fc4_sigma', 
                                 activation_fn=None)
        eps = tf.random_normal(
            shape=tf.shape(self.z_log_sigma_sq),
            mean=0, stddev=1, dtype=tf.float32)
        self.z = self.z_mu + tf.sqrt(tf.exp(self.z_log_sigma_sq)) * eps

        # Decode
        # z -> x_hat
        g1 = fc(self.z, 16, scope='dec_fc1', activation_fn=tf.nn.relu)
        g2 = fc(g1, 32, scope='dec_fc2', activation_fn=tf.nn.relu)
        g3 = fc(g2, 64, scope='dec_fc3', activation_fn=tf.nn.relu)
#         self.x_hat = fc(g3, input_dim, scope='dec_fc4', 
#                         activation_fn=tf.sigmoid)
        self.x_hat = fc(g3, input_dim, scope='dec_fc4', 
                        activation_fn=None)
        # Loss
        # Reconstruction loss
        # Minimize the cross-entropy loss
        # H(x, x_hat) = -\Sigma x*log(x_hat) + (1-x)*log(1-x_hat)
        epsilon = 1e-10
#         recon_loss = -tf.reduce_sum(
#             self.x * tf.log(epsilon+self.x_hat) + 
#             (1-self.x) * tf.log(epsilon+1-self.x_hat), 
#             axis=1
#         )
        recon_loss = tf.square(self.x-self.x_hat)
        self.recon_loss = tf.reduce_mean(recon_loss)

        # Latent loss
        # KL divergence: measure the difference between two distributions
        # Here we measure the divergence between 
        # the latent distribution and N(0, 1)
        latent_loss = -0.5 * tf.reduce_sum(
            1 + self.z_log_sigma_sq - tf.square(self.z_mu) - 
            tf.exp(self.z_log_sigma_sq), axis=1)
        self.latent_loss = tf.reduce_mean(latent_loss)

        self.total_loss = self.recon_loss*self.recon_coef + self.latent_loss
        self.train_op = tf.train.AdamOptimizer(
            learning_rate=self.learning_rate).minimize(self.total_loss)
        
        self.losses = {
            'recon_loss': self.recon_loss,
            'latent_loss': self.latent_loss,
            'total_loss': self.total_loss,
        }        
        return

    # Execute the forward and the backward pass
    def run_single_step(self, x):
        _, losses = self.sess.run(
            [self.train_op, self.losses],
            feed_dict={self.x: x}
        )
        return losses

    # x -> x_hat
    def reconstructor(self, x):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.x: x})
        return x_hat

    # z -> x
    def generator(self, z):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.z: z})
        return x_hat
    
    # x -> z
    def transformer(self, x):
        z = self.sess.run(self.z, feed_dict={self.x: x})
        return z
    
    def transformer_mu(self, x):
        z = self.sess.run(self.z_mu, feed_dict={self.x: x})
        return z    
```

```python
class VariantionalAutoencoder_near(object):

    def __init__(self, learning_rate=1e-4, batch_size=64, n_z=16, recon_coef = 1):
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.n_z = n_z
        self.recon_coef = recon_coef

        tf.reset_default_graph()
        self.build()

        self.sess = tf.InteractiveSession()
        self.sess.run(tf.global_variables_initializer())

    def build_encoder(self, input_x, name, reuse=False):    
        with tf.variable_scope(name, reuse=reuse):
            # Encode
            # x -> z_mean, z_sigma -> z
            f1 = fc(input_x, 256, scope='enc_fc1', activation_fn=tf.nn.relu)
            f2 = fc(f1, 128, scope='enc_fc2', activation_fn=tf.nn.relu)
            f3 = fc(f2, 64, scope='enc_fc3', activation_fn=tf.nn.relu)
            z_mu = fc(f3, self.n_z, scope='enc_fc4_mu', 
                           activation_fn=None)
            z_log_sigma_sq = fc(f3, self.n_z, scope='enc_fc4_sigma', 
                                     activation_fn=None)
            eps = tf.random_normal(
                shape=tf.shape(z_log_sigma_sq),
                mean=0, stddev=1, dtype=tf.float32)
            z = z_mu + tf.sqrt(tf.exp(z_log_sigma_sq)) * eps
        
        return z_mu, z_log_sigma_sq, z

    def build_decoder(self, input_z, name, reuse=False):    
        with tf.variable_scope(name, reuse=reuse):
#             de_in = tf.concat((self.z, self.z_next), axis=1)
#             print('shape',self.z.shape, self.x.shape, de_in.shape)
            g1 = fc(input_z, 64, scope='dec_fc1', activation_fn=tf.nn.relu)
            g2 = fc(g1, 128, scope='dec_fc2', activation_fn=tf.nn.relu)
            g3 = fc(g2, 256, scope='dec_fc3', activation_fn=tf.nn.relu)
    #         self.x_hat = fc(g3, self.input_dim, scope='dec_fc4', 
    #                         activation_fn=tf.sigmoid)
            x_hat = fc(g3, self.input_dim*2, scope='dec_fc4', 
                            activation_fn=None)
        
        return x_hat
        
    # Build the netowrk and the loss functions
    def build(self):
        self.input_dim = 2
        self.x = tf.placeholder(
            name='x', dtype=tf.float32, shape=[None, self.input_dim])
        self.x_next = tf.placeholder(
            name='x_next', dtype=tf.float32, shape=[None, self.input_dim])

        input_x = tf.concat((self.x, self.x_next), axis=1)
        self.z_mu, self.z_log_sigma_sq, self.z = self.build_encoder(input_x, 'x')
        
#         self.z_mu_x, self.z_mu = tf.split(self.z_mu_ori, [2, 1], 1)
#         self.z_log_sigma_sq_x, self.z_log_sigma_sq = tf.split(self.z_log_sigma_sq_ori, [2, 1], 1)
#         self.z_x, self.z = tf.split(self.z_ori, [2, 1], 1)
    
        self.x_hat = self.build_decoder(self.z, 'decoder')
        # Loss
        # Reconstruction loss
        # Minimize the cross-entropy loss
        # H(x, x_hat) = -\Sigma x*log(x_hat) + (1-x)*log(1-x_hat)
        epsilon = 1e-10
#         recon_loss = -tf.reduce_sum(
#             self.x * tf.log(epsilon+self.x_hat) + 
#             (1-self.x) * tf.log(epsilon+1-self.x_hat), 
#             axis=1
#         )
#         x_pair = self.x_next - self.x #tf.concat((self.x, self.x_next), axis=1)
        recon_loss = tf.square(input_x-self.x_hat)
        self.recon_loss = tf.reduce_mean(recon_loss)

        # Latent loss
        # KL divergence: measure the difference between two distributions
        # Here we measure the divergence between 
        # the latent distribution and N(0, 1)
        latent_loss = -0.5 * tf.reduce_sum(
            1 + self.z_log_sigma_sq - tf.square(self.z_mu) - 
            tf.exp(self.z_log_sigma_sq), axis=1)
        self.latent_loss = tf.reduce_mean(latent_loss)

        self.total_loss = self.recon_loss * self.recon_coef + self.latent_loss
        self.train_op = tf.train.AdamOptimizer(
            learning_rate=self.learning_rate).minimize(self.total_loss)
        
        self.losses = {
            'recon_loss': self.recon_loss,
            'latent_loss': self.latent_loss,
            'total_loss': self.total_loss,
        }        
        return

    # Execute the forward and the backward pass
    def run_single_step(self, x, x_next):     
        _, losses = self.sess.run(
            [self.train_op, self.losses],
            feed_dict={self.x: x, self.x_next: x_next}
        )
        return losses

    # x -> x_hat
    def reconstructor(self, x, x_next):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.x: x, self.x_next: x_next})
        return x_hat

    # z -> x
    def generator(self, z, z_next):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.z:z, self.z_next:z_next})
        return x_hat
    
    # x -> z
    def transformer(self, x, x_next):
        z = self.sess.run([self.z], feed_dict={self.x: x, self.x_next: x_next})
        return z
    
    def transformer_mu(self, x, x_next):
        z = self.sess.run([self.z_mu], feed_dict={self.x: x, self.x_next: x_next})
        return z
```

```python
def trainer(model_object, data, learning_rate=1e-4, 
            batch_size=64, num_epoch=100, n_z=2, log_step=50, recon_coef=1):
    model = model_object(
        learning_rate=learning_rate, batch_size=batch_size, n_z=n_z, recon_coef=recon_coef)
    
    data_index = np.arange(len(data))
    print(len(data), data.shape)
    
    for epoch in range(num_epoch):
        start_time = time.time()
        np.random.shuffle(data_index)
        for iter in range(0, data_size, batch_size):
            # Get a batch
            inds = data_index[iter:iter+batch_size]
            batch = data[inds]
            losses = model.run_single_step(batch)
        end_time = time.time()
        
        if epoch % log_step == 0:
            log_str = '[Epoch {}] '.format(epoch)
            for k, v in losses.items():
                log_str += '{}: {:.3f}  '.format(k, v)
            log_str += '({:.3f} sec/epoch)'.format(end_time - start_time)
            print(log_str)
            
    print('Done!')
    return model
```

```python
def trainer_near(model_object, data, data_next, learning_rate=1e-4, 
            batch_size=64, num_epoch=100, n_z=2, log_step=50, recon_coef=1):
    model = model_object(
        learning_rate=learning_rate, batch_size=batch_size, n_z=n_z, recon_coef=recon_coef)
    
    data_index = np.arange(len(data))
    print(len(data))
    
    for epoch in range(num_epoch):
        start_time = time.time()
        np.random.shuffle(data_index)
        for iter in range(0, len(data), batch_size):
            # Get a batch
            inds = data_index[iter:iter+batch_size]
            batch, batch_next = data[inds], data_next[inds]
            losses = model.run_single_step(batch, batch_next)
        end_time = time.time()
        
        if epoch % log_step == 0:
            log_str = '[Epoch {}] '.format(epoch)
            for k, v in losses.items():
                log_str += '{}: {:.3f}  '.format(k, v)
            log_str += '({:.3f} sec/epoch)'.format(end_time - start_time)
            print(log_str)
            
    print('Done!')
    return model
```
